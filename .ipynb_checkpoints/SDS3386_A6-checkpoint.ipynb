(c) Perform a Principal Component Analysis of the entire standardized dataset, and produce figures as follows:

    i. PC1 score vs. PC2 score, with colour indicating the object label.
    ii. PC3 score vs. PC4 score, with colour indicating the object label.
    iii. A figure showing the first 8 principal components.

Hint: If using Scikit-learn, the Principal Components are returned as pca.components_, while the PC scores are the result of applying pca.transform.

```python
from sklearn.decomposition import PCA
import altair as alt

k = 100
pca = PCA(n_components=k).fit(X)
transformed_X = pca.transform(X)

df_pca = pd.DataFrame(transformed_X, columns = [f'PC {i}' for i in range(k)])

# (i) and (ii)
chart_i = alt.Chart(df_pca).mark_point().encode(
    x = alt.X("PC 1", title = "PC1 Score"),
    y = alt.X("PC 2", title = "PC2 Score")
).properties(title = "PC1 score vs. PC2 score")

chart_ii = alt.Chart(df_pca).mark_point().encode(
    x = alt.X("PC 3", title = "PC3 Score"),
    y = alt.Y("PC 4", title = "PC4 Score")
).properties(title = "PC3 score vs. PC4 score")

chart_i | chart_ii
```

```python
# (iii) 

figure_images = plt.figure(figsize=(10,5))
components = pca.components_

figure_images.add_subplot(numRows, numColumns, 1)
plt.imshow(components[0].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 1")

figure_images.add_subplot(numRows, numColumns, 2)
plt.imshow(components[1].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 2")

figure_images.add_subplot(numRows, numColumns, 3)
plt.imshow(components[2].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 3")

figure_images.add_subplot(numRows, numColumns, 4)
plt.imshow(components[3].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 4")

figure_images.add_subplot(numRows, numColumns, 5)
plt.imshow(components[4].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 5")

figure_images.add_subplot(numRows, numColumns, 6)
plt.imshow(components[5].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 6")

figure_images.add_subplot(numRows, numColumns, 7)
plt.imshow(components[6].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 7")

figure_images.add_subplot(numRows, numColumns, 8)
plt.imshow(components[7].reshape(32,32), cmap = plt.cm.bone)
plt.axis('off')
plt.title("Component 8")
```

## Question 2 (5 marks)

This question uses the dataset SAheart.csv, as in the Week 9 Lecture.

(a) Convert the famhist column into a binary variable with values 0 and 1.


```python
sa_heart_df = pd.read_csv("SAHeart.csv")
sa_heart_df['famhist'] = sa_heart_df['famhist'].map({'Present':1, 'Absent':0})
```

(b) Randomly split the dataset into training and test datasets, with the training dataset containing approximately 2/3 of the data. Set random_state=1.


```python
from sklearn.model_selection import train_test_split

X = sa_heart_df.drop(columns=['chd']) #In the next question, we will use the variables kept to predict chd
y = sa_heart_df['chd']


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=2/3, random_state = 1)
```

(c) Use logistic regression to predict the variable chd from the others. Report the accuracy and log loss.




```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss

logistic_model = LogisticRegression(random_state = 1) 
logistic_model.fit(x_train, y_train)

y_pred = logistic_model.predict(x_test)

print("The accuracy rate is:", accuracy_score(y_test, y_pred) * 100, "%")
print("The log loss is:", log_loss(y_test, y_pred))
```

(d) Apply PCA to the training set, produce a scree plot, and choose a number of components K to retain, with 1 < K < 9. Thereâ€™s no â€œright answerâ€ here.)


```python
pca = PCA()
pca_training_x = pca.fit_transform(x_train)

explained_var_ratio = pca.explained_variance_ratio_ 

num_of_components = len(explained_var_ratio)
plt.figure(figsize=(5, 5))
plt.plot(np.arange(1, num_of_components + 1), explained_var_ratio, marker='o', linestyle='-')
plt.xlabel('Num. of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

(e) Compute the first K PC scores of your test set (using the PCs calculated above from the training set).


```python
k = 10

pca_test_x = pca.transform(x_test)[:, :k]
```

(f) Use these PC scores to predict, on the test set, the variable chd from the others, using logistic regression. Report the accuracy and log loss.

```python
PC_logistic_model = LogisticRegression(random_state = 1)

PC_logistic_model.fit(pca_training_x[:, :k], y_train)

y_pred = PC_logistic_model.predict(pca_test_x)

print("The accuracy rate is:", accuracy_score(y_test, y_pred) * 100, "%")
print("The log loss is:", log_loss(y_test, y_pred))
```
