## Question 2 (4 marks)

(a) For each sex separately, compute the sample mean and variance of flipper length. We model flipper length using a univariate normal distribution. Specifically, we suppose that flipper lengths of males follow a normal distribution $N(\mu_M, \sigma^2_M)$, and we fit the model to the data by setting the parameters $\mu_M$ and $\sigma^2_M$ equal to the sample mean and variance computed above. We do the same for the females. We plot the probability density functions of these two fitted models, with colours approximately matching the histograms from Question 1. We use `scipy.stats.norm.pdf` and any standard graphics package.

```python
mean_flipper_F = adelie_df[adelie_df['sex'] == 'FEMALE']['flipper_length_mm'].mean()
var_flipper_F = adelie_df[adelie_df['sex'] == 'FEMALE']['flipper_length_mm'].var()

mean_flipper_M = adelie_df[adelie_df['sex'] == 'MALE']['flipper_length_mm'].mean()
var_flipper_M = adelie_df[adelie_df['sex'] == 'MALE']['flipper_length_mm'].var()


print(f'The Mean flipper length of Female Adelie is: {mean_flipper_F:.2f}; and similarly, the Female Variance is: {var_flipper_F:.2f}')
print(f'The Mean flipper length of Male Adelie is: {mean_flipper_M:.2f}; and similarly, the Male Variance is: {var_flipper_M:.2f}')

x1 = np.arange(150, 210, step=0.05)
y1 = stats.norm.pdf(x1, loc=mean_flipper_F, scale=var_flipper_F**0.5)
pdf_1 = pd.DataFrame(zip(x1,y1), columns=['flipper_length_mm', 'pdf'])
p1 = alt.Chart(pdf_1).mark_point(color = 'blue', size=2).encode(
    x = alt.X('flipper_length_mm', title = 'Flipper Length (mm)').scale(zero=False),
    y = alt.Y('pdf')
    
).interactive()

x2 = np.arange(150, 210, step=0.05)
y2 = stats.norm.pdf(x2, loc=mean_flipper_M, scale=var_flipper_M**0.5)
pdf_2 = pd.DataFrame(zip(x2,y2), columns=['flipper_length_mm', 'pdf'])
p2 = alt.Chart(pdf_2).mark_point(color = 'orange', size=2).encode(
    x = alt.X('flipper_length_mm', title = 'Flipper Length (mm)').scale(zero=False),
    y = alt.Y('pdf')
).interactive()

p1 + p2
```

(b) Suppose that Adelie penguins are known to be approximately 50% male and 50% female. This is our prior belief about the sex of any new penguin we consider, before we observe anything about them. We observe a new penguin, “Bela” and measure their flipper length to be 200mm. Using your models from part (a), compute $$p(flipper length == 200 | sex == male)$$ $$p(flipper length == 200 | sex == female)$$

Use Bayes’ rule to compute your posterior belief about Bela’s sex. (If you prefer, you can do part (c) first, and use that to answer this part.)



From part (C) we get that  $$p(flipper length == 200 | sex == male) = 0.8254$$ $$p(flipper length == 200 | sex == female) = 0.1746$$

(c) Write a function bayes_classify_1() that automates the calculation in part (b). It should take as inputs: a flipper length, and the means and variances you computed in part (a); and give as output a vector of length two containing: the posterior probability that the penguin is female, and the posterior probability that the penguin is male. You may use, with modification, a function from the Week 5 Lab.

```python
def bayes_classifiy_1(x, means, variances):
    
    likelihood_0 = stats.norm.pdf(x, loc=means[0], scale=variances[0]**0.5) #Square root variance -> std
    likelihood_1 = stats.norm.pdf(x, loc=means[1], scale=variances[1]**0.5) #pdf only takes mean and std (not variance)
    
    evidence = likelihood_0 * 0.5 + likelihood_1 * 0.5
    posterior_0 = (likelihood_0 / evidence) * 0.5
    posterior_1 = (likelihood_1 / evidence) * 0.5

    posteriors = [posterior_0, posterior_1]
    
    return posteriors

bayes_classifiy_1(200, [mean_flipper_F, mean_flipper_M], [var_flipper_F, var_flipper_M])
```

(d) Apply your classifier to all of the Adelie penguins in the dataset and produce a chart showing flipper length, actual sex and predicted sex, where “predicted sex” is the sex with the highest posterior probability according to the classifier. 

```python
adelie_classifier = bayes_classifiy_1(adelie_df['flipper_length_mm'], [mean_flipper_F, mean_flipper_M], [var_flipper_F, var_flipper_M])

predicted_sex = []

for i in range(len(adelie_df)):
    if(adelie_classifier[0][i] > adelie_classifier[1][i]):
        predicted_sex.append('FEMALE')
    else:
        predicted_sex.append('MALE')
        
adelie_df['predicted_sex'] = predicted_sex

prediction_chart = alt.Chart(adelie_df).encode(
    x = alt.X('flipper_length_mm', title = "Flipper Length (mm)").scale(domain=(140, 220)),
    y = alt.Y('sex'),
    color = alt.Color('predicted_sex', title = 'Predicted Sex')
).mark_point().properties(title = "Predicted Sex vs. Actual Sex (Adelie)").interactive()

prediction_chart
```

## Question 3 (4 marks)

(a) Create a training set consisting of the first 100 Adelie penguins in the original dataset. Create a test set consisting of the remaining ones. Using the method from Question 2, fit your model (i.e. calculate the means and variances) using the training set only, and then use your bayes_classify_1 function to classify all the penguins in the test set. (This is a probabilistic classifier.) For each penguin, define the predicted sex as the one with the highest posterior probability; this is the associated deterministic classifier. Calculate the accuracy of the deterministic classifier on the test set, and calculate the mean log loss of the probablistic classifier on the test set. (For mean log loss, you can either use the definition directly, or use a function from a standard package.)

```python
training_set = adelie_df.iloc[:100, ]
test_set = adelie_df.iloc[100:, ]

training_female = training_set[training_set['sex'] == 'FEMALE']
mean_female = training_female['flipper_length_mm'].mean()
var_female = training_female['flipper_length_mm'].var()

training_male = training_set[training_set['sex'] == 'MALE']
mean_male = training_male['flipper_length_mm'].mean()
var_male = training_male['flipper_length_mm'].var()

adelie_classifier_2 = bayes_classifiy_1(test_set['flipper_length_mm'], [mean_female, mean_male], [var_female, var_male])

predicted_sex_test = []

for i in range(len(test_set)):
    if(adelie_classifier_2[0][i] > adelie_classifier_2[1][i]):
        predicted_sex_test.append('FEMALE')
    else:
        predicted_sex_test.append('MALE')
        
actual_sex = test_set['sex'].to_list()
correct_classification = 0

for i in range(len(actual_sex)):
    if predicted_sex_test[i] == actual_sex[i]:
        correct_classification += 1

accuracy = correct_classification / len(predicted_sex_test)

log_loss = []


for i in range(len(actual_sex)):
    if actual_sex[i] == 'FEMALE':
        log_loss.append(-math.log(adelie_classifier_2[0][i]))
    else:
        log_loss.append(-math.log(adelie_classifier_2[1][i]))
        
mean_log_loss = sum(log_loss) / len(actual_sex)

test_set['predicted_sex'] = predicted_sex_test 

print("The accuracy of the deterministic classifier on the test set is: ", accuracy*100, "%")
print("The mean log loss of the probablistic classifier on the test set is: ", mean_log_loss*100, "%")
```

(b) Redo the classification using a Naive Bayes classifier that uses both flipper lengths and bill lengths. Write your own function naive_bayes_classify that generalizes your bayes_classify_1 function. It should take as inputs: a vector of predictor variables, two vectors of means (one for each class, e.g. male or female), and two vectors of variances, and output a vector of length 2 containing the posterior probabilities for each class. Using the same training and test sets as in part (a), fit your models using the training data only, classify all penguins in the test set, and then calculate the accuracy and mean log loss on the test set. Note: do not use a Naïve Bayes classifier from an existing package.

```python
def naive_bayes_classifier(predictors, female_means, male_means, female_vars, male_vars):
    likelihood_female_0 = stats.norm.pdf(predictors[0], loc=female_means[0], scale=female_vars[0]**0.5)
    likelihood_female_1 = stats.norm.pdf(predictors[1], loc=female_means[1], scale=female_vars[1]**0.5)
    
    likelihood_male_0 = stats.norm.pdf(predictors[0], loc=male_means[0], scale=male_vars[0]**0.5)
    likelihood_male_1 = stats.norm.pdf(predictors[1], loc=male_means[1], scale=male_vars[1]**0.5)
    
    likelihood_female_total = likelihood_female_0 * likelihood_female_1
    likelihood_male_total = likelihood_male_0 * likelihood_male_1
    
    evidence = likelihood_female_total * 0.5 + likelihood_male_total * 0.5
    
    posterior_female = (likelihood_female_total / evidence) * 0.5
    posterior_male = (likelihood_male_total / evidence) * 0.5
    
    return [posterior_female, posterior_male]


mean_flipper_F = training_female['flipper_length_mm'].mean()
mean_bill_F = training_female['bill_length_mm'].mean()

var_flipper_F = training_female['flipper_length_mm'].var()
var_bill_F = training_female['bill_length_mm'].var()

mean_flipper_M = training_male['flipper_length_mm'].mean()
mean_bill_M = training_male['bill_length_mm'].mean()

var_flipper_M = training_male['flipper_length_mm'].var()
var_bill_M = training_male['bill_length_mm'].var()


adelie_classifier_3 = naive_bayes_classifier([test_set['flipper_length_mm'], test_set['bill_length_mm']], 
                                             [mean_flipper_F, mean_bill_F], [mean_flipper_M, mean_bill_M],
                                             [var_flipper_F, var_bill_F], [var_flipper_M, var_bill_M])

predicted_sex_test_2 = []

for i in range(len(test_set)):
    if adelie_classifier_3[0][i] > adelie_classifier_3[1][i]:
        predicted_sex_test_2.append('FEMALE')
    else:
        predicted_sex_test_2.append('MALE')
    
    
correct_classification = 0

for i in range(len(actual_sex)):
    if predicted_sex_test_2[i] == actual_sex[i]:
        correct_classification += 1

accuracy_naive_bayes = correct_classification / len(predicted_sex_test_2)

log_loss_2 = []


for i in range(len(actual_sex)):
    if actual_sex[i] == 'FEMALE':
        log_loss_2.append(-math.log(adelie_classifier_3[0][i]))
    else:
        log_loss_2.append(-math.log(adelie_classifier_3[1][i]))
        
mean_log_loss_naive_bayes = sum(log_loss_2) / len(actual_sex)

test_set['predicted_sex_naive_bayes'] = predicted_sex_test_2

print("The accuracy of the naive bayes classifier on the test set is: ", accuracy_naive_bayes*100, "%")
print("The mean log loss of the naive bayes classifier on the test set is: ", mean_log_loss_naive_bayes*100, "%")
```
