## Question 2 (2 marks)

Download this dataset: <https://github.com/mwaskom/seaborn-data/blob/master/penguins.csv> .

Produce one or more bar charts, using Altair (also called Vega-Altair), showing how many male and female penguins of each species are in the dataset. Note that for some penguins, no sex is specified; represent these as well. Choose a chart type that allows viewers to easily compare all of the numbers represented.

*In this and the following questions, there is no single right answer, but you should aim for a high-quality visualization suitable for use in a report. In particular, all charts should have appropriate titles and axis labels.*

Hint: "null" and "Count of Records" are not suitable labels for a report.

```python
#Import pandas -> loading csv file in a python-dataframe | Import altair -> graph visualization
import pandas as pd  
import altair as alt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv" #Grabbing raw data from github csv file
seaborn_df = pd.read_csv(url) #Converting csv -> pandas dataframe

seaborn_df['sex'] = seaborn_df['sex'].fillna('UNKNOWN') #Filling in missing values for gender

adelie_chart = alt.Chart(seaborn_df).encode(
    x='sex',
    y=alt.Y('count()',title='Total # of Adelie').scale(domain=(0,80)), 
    color=alt.Color('sex').scale(scheme='pastel2')
).mark_bar().transform_filter(alt.datum.species == 'Adelie').properties(width=200, title = ('Adelie vs. Gender'))

chinstrap_chart = alt.Chart(seaborn_df).encode(
    x='sex',
    y=alt.Y('count()',title='Total # of Chinstrap').scale(domain=(0,80)),
    color=alt.Color('sex').scale(scheme='pastel2')
).mark_bar().transform_filter(alt.datum.species == 'Chinstrap').properties(width=200, title = ('Chinstrap vs. Gender'))

gentoo_chart = alt.Chart(seaborn_df).encode(
    x='sex',
    y=alt.Y('count()',title='Total # of Gentoo').scale(domain=(0,80)),
    color=alt.Color('sex').scale(scheme='pastel2')
).mark_bar().transform_filter(alt.datum.species == 'Gentoo').properties(width=200, title = ('Gentoo vs. Gender'))



adelie_chart | chinstrap_chart | gentoo_chart
```

a) There are several reasons why the above chart is poorly designed: 
- The graph legend is unclear. What is size and what does 1, 2, ..., 6 imply? What unit of measurement is size? 
- The color spectrum used for 1-6 is not optimal for this graph. An increasing quantitative measurement from 1 to 6 should ideally correspond with progressing from a lighter shade to darker, vice-versa, or other thoughtfully chosen color spectrums
- The name in the y-axis is confusing. As an alternative, the name of it should be "Collective # of tips per gender" for better clarity
- The tipping percentage is not taken into consideration. For example, what if the total bill was expensive and the tipping was little, compared to an inexpensive total bill with a large amount tipped
- This graph does not take into consideration the time of tipping which is one of the data points given, "time"

```python
# Part b)
import pandas as pd
import altair as alt

tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

#Before plotting the graph, let's collect one-dimensional statistical information from the dataframe
print("Number of Female: ", tips['sex'].value_counts()['Female']) #87
print("Number of Male: ", tips['sex'].value_counts()['Male']) #157

print("Lunch Tips: ", tips['time'].value_counts()['Lunch']) #68
print("Dinner Tips: ", tips['time'].value_counts()['Dinner']) #176

#From the above data, we can derive two initial conclusions: 
#1. There are more data points for male tipping customers than female tipping customers, thus skewing the data towards males
#   As a result, it would be invalid to display a graph using sum() or count() of tipping data as there are only half as many female tipping datapoints
#2. People tip more during dinner compared to lunch

#Now, let us try to fix one significant flaw of the previous graph: The tipping amount is NOT standardized across the datapoints
# i.e. If total bill was expensive -> tipping was little, compared to inexpensive total bill -> large amount tipped
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100


improved_chart = alt.Chart(tips).encode(
    x=alt.X('tip_percentage',title='Tipping % of Customers'),
    y=alt.Y('sex'),
    color=alt.Color('sex').scale(scheme='set2'),
    shape = 'time',
    facet = 'smoker',
).mark_boxplot(size = 70).properties(width = 300, height = 200)

improved_chart
```

b) Above we have plotted the relation between tip_percentage and sex using a box-plot. The reason for choosing a box-plot over a scatter plot is because the data is skewed towards males (there are nearly double the amount of tipping data of males than females, therefore not representative of the female population). Using functions like sum() or count() would simply aggregate an excess number of tipping data from males, and not show the % of tip contributed by the individual genders. Hence, this is another reason why the above plot is an improvement over the previous graph. In addition, the size column is not used for the graph due to its ambigiuity (what does size represent?).

c) Here are the results for the median tip_percentage for the 4 following groups: 
- Female + non-smoker = 14.96
- Female + smoker = 17.39
- Male + non-smoker = 15.76
- male + smoker = 14.10

We can therefore conclude from the standardization of data above, that female smokers have the highest tipping percentage out of any group. 

Another inference we can make is people tend to tip more, or outside the interquartile range during dinner time, compared to lunch based on the outlier tips in the graph.
